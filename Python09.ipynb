### 9.1.2. Presentation - Artificial Neural Network

> **Mathematical model:**
>  * $x=\sum_{i=1}^{I} X_{i} W_{i}^+B_{i}$

>  * $Y=f(x)=f\left(\sum_{i=1}^{I} X_{i} W_{i}+B_{i}\right)$

> <img src=	"	https://i.ibb.co/hcjZVDk/9-3.png	"	width="400"/>

> **Optimization:**
>  * What are the optimum values of weights and biases such minimize (maximize) error (accuracy) measurement of estimated Y versus actual Y?
>  * We pick an optimizer (to be seen in the future) to optimize our weights and biases

### 9.1.5. Presentation - Neural Network Optimization

> **Minimize a so-called loss function (also referred to as cost function):**
> * Example: $l(p)=\frac{1}{2 N} \sum_{n=1}^{N}\left\|\boldsymbol{Y}_{n}-\tilde{\boldsymbol{Y}}_{n}\right\|^{2}$
>  * The term p is a set of optimization variables such as weights and biases of different layers 
>  * The terms $\boldsymbol{Y}_{n}$ and $\tilde{\boldsymbol{Y}}_{n}$ are the actual and estimated outputs of the $n^{th}$ training datapoint

> <img src=	"	https://i.ibb.co/8snKxvy/9-4.png	"	width="500"/>

> **Optimization/learning process:**
> * Iterative optimization
>  * Feed forward
>  * Error measurements
>  * Backpropagation
> * Optimizers
>  * Gradient descent 
>  * Adaptive Moment Estimation (Adam)

### 9.1.6. Presentation - Stochastic Learning
> **Divide training repository into batches of training datapoints:**
> * Memory-friendly batch-size
> * Memory friendly batch-number 

> **Stochastic training:**
> * feed each batch to the network one at a time and update $p$
> * Once all the batches pass the network, one iteration is counted
> * Train for a number of iterations
> * Example: 32 batches each with 32 datapoints (total number of data points $N=1000$)
